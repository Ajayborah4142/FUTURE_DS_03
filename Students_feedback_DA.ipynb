# *2) Data Analysis % Find Insights* 

###  Q.1)	What is the overall average rating for each feedback parameter?

```python
# Step 3: Drop unnecessary columns (like index or ID columns)
data     = data.drop(columns=['Unnamed: 0', 'Student ID'], errors='ignore')

# Step 4: Calculate overall average rating for each feedback parameter
average_ratings = data.mean().round(2)
average_ratings
```

```python
## Visualize 

average_ratings.plot(kind='bar', color='skyblue')
plt.title('Average Ratings by Course Aspect')
plt.ylabel('Average Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Insights

* Instructor Strength: Highly knowledgeable (7.50) — students recognize strong subject expertise.

* Moderate Areas: Explanation clarity (6.08) and course structure (5.64) are average — teaching delivery and organization could be improved.

* Weak Areas: Assignments (5.43) and doubt-solving (5.47) need attention — students may find tasks difficult and support limited.

* Overall Insight: While the instructor is well-versed in the subject, improving clarity, support, and engagement methods would enhance overall learning satisfaction.

### Q.2)  Which aspect of the course received the highest and lowest average ratings?

```python
highest_aspect = average_ratings.idxmax()
lowest_aspect = average_ratings.idxmin()

print(f"Highest Rated Aspect: {highest_aspect} = {average_ratings[highest_aspect]:.2f}")
print(f"Lowest Rated Aspect: {lowest_aspect} = {average_ratings[lowest_aspect]:.2f}")
```

### Insights

* Strength: Students rated the instructor as well-versed in the subject (7.50/10), indicating strong knowledge and teaching capability.

* Area for Improvement: The degree of difficulty of assignments scored lower (5.43/10), suggesting assignments may be too challenging or not aligned with learning pace.

###  Q.3)  What is the overall satisfaction level of students (average of all ratings per student)?

```python
# Columns in your dataset:
rating_columns = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]

#  Step 1: Calculate per-student average rating
data['average_rating'] = data[rating_columns].mean(axis=1)

#  Step 2: Calculate overall satisfaction level
overall_satisfaction = data['average_rating'].mean()

print(" Overall Satisfaction Level of Students:", round(overall_satisfaction, 2))
```

### Q.7)   Which feedback parameter shows the highest variation (most disagreement among students)?

```python
feedback_columns = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]
```

### Q.8)  How many students gave an overall rating above 8 (indicating high satisfaction)?

```python
# List of feedback columns
feedback_columns = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]

# Calculate overall rating per student (average across all feedback columns)
data['Overall_rating'] = data[feedback_columns].mean(axis=1)

# Count students with overall rating above 8
high_satisfaction_count = data[data['Overall_rating'] > 8].shape[0]

print("Number of students with high satisfaction (rating > 8):", high_satisfaction_count)
```

### Q.9)  What are the top 3 factors that most influence “Course recommendation based on relevance”?

```python
# Feedback columns excluding the target
features = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond'
]

# Calculate correlation with course recommendation
correlations = data[features + ['Course recommendation based on relevance']].corr()
target_corr = correlations['Course recommendation based on relevance'].drop('Course recommendation based on relevance')
top_3_factors = target_corr.abs().sort_values(ascending=False).head(3)

print("Top 3 factors influencing Course Recommendation:")
print(top_3_factors)
```

### Q.10)  Are there any students who consistently gave very low or very high ratings across all parameters?


```python
# Feedback columns
feedback_columns = [
    'Well versed with the subject',
    'Explains concepts in an understandable way',
    'Use of presentations',
    'Degree of difficulty of assignments',
    'Solves doubts willingly',
    'Structuring of the course',
    'Provides support for students going above and beyond',
    'Course recommendation based on relevance'
]

# Students with all very high ratings
very_high_students = data[(data[feedback_columns] >= 9).all(axis=1)]

# Students with all very low ratings
very_low_students = data[(data[feedback_columns] <= 4).all(axis=1)]

print("Students with consistently very high ratings:", very_high_students.shape[0])
print("Students with consistently very low ratings:", very_low_students.shape[0])
```
